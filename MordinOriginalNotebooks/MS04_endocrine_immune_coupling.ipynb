```python
class EndocrineImmuneField:
    """Models the unified endocrine-immune field interactions."""
    
    def __init__(self, size=256, device='cuda'):
        self.size = size
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        
        # Base field
        self.field = torch.zeros(1, 1, size, size, dtype=torch.complex64, device=self.device)
        
        # Endocrine glands with their base frequencies
        self.glands = {
            'pineal': {'pos': (0.0, 0.85), 'freq': 40.0, 'hormone': 'melatonin'},
            'pituitary': {'pos': (0.0, 0.75), 'freq': 20.0, 'hormone': 'master control'},
            'thyroid': {'pos': (0.0, 0.6), 'freq': 15.0, 'hormone': 'metabolism'},
            'thymus': {'pos': (0.0, 0.4), 'freq': 25.0, 'hormone': 'T-cells'},
            'adrenals': {'pos': (-0.3, 0.0), 'freq': 10.0, 'hormone': 'stress'},
            'pancreas': {'pos': (0.1, 0.1), 'freq': 8.0, 'hormone': 'insulin'},
            'gonads': {'pos': (0.0, -0.5), 'freq': 5.0, 'hormone': 'sex hormones'}
        }
        
        # Immune centers with their frequencies
        self.immune_centers = {
            'bone_marrow': {'pos': [(0.3, 0.2), (-0.3, -0.2)], 'freq': 30.0, 'cells': 'all blood cells'},
            'spleen': {'pos': (-0.4, 0.15), 'freq': 18.0, 'cells': 'B-cells, filtering'},
            'lymph_nodes': {'pos': [(0.2, 0.5), (-0.2, 0.3), (0.0, -0.3)], 'freq': 22.0, 'cells': 'lymphocytes'},
            'peyers_patches': {'pos': (0.0, -0.1), 'freq': 12.0, 'cells': 'gut immunity'}
        }
        
        # Coupling strengths between systems
        self.coupling_matrix = self.create_coupling_matrix()
        
        # Time-dependent hormone levels
        self.hormone_levels = {gland: 0.5 for gland in self.glands}
        self.cytokine_levels = {center: 0.3 for center in self.immune_centers}
        
        # Circadian phase
        self.circadian_phase = 0.0
        
        # Disease/stress perturbations
        self.stress_level = 0.0
        self.infection_sites = []
        
    def create_coupling_matrix(self):
        """Create the coupling strengths between endocrine and immune systems."""
        # Key couplings based on real physiology
        couplings = {
            ('thymus', 'bone_marrow'): 0.8,      # T-cell development
            ('adrenals', 'lymph_nodes'): -0.6,    # Cortisol suppresses immunity
            ('thyroid', 'bone_marrow'): 0.5,      # Metabolism affects cell production
            ('pituitary', 'all'): 0.4,            # Master regulator
            ('pineal', 'thymus'): 0.7,            # Melatonin boosts immunity
            ('gonads', 'thymus'): -0.3,           # Sex hormones vs thymus
        }
        return couplings
    
    def initialize_gland_fields(self):
        """Create initial resonance patterns for all glands."""
        x = torch.linspace(-1, 1, self.size, device=self.device)
        y = torch.linspace(-1, 1, self.size, device=self.device)
        X, Y = torch.meshgrid(x, y, indexing='ij')
        
        # Endocrine glands
        for gland, info in self.glands.items():
            gx, gy = info['pos']
            # Gaussian envelope with oscillation
            envelope = torch.exp(-10 * ((X - gx)**2 + (Y - gy)**2))
            phase = 2 * np.pi * info['freq'] * 0.1  # Initial phase
            
            # Hormone level modulates amplitude
            amplitude = self.hormone_levels[gland]
            # Create complex exponential using cos + i*sin
            self.field += amplitude * envelope * torch.complex(torch.cos(torch.tensor(phase)), torch.sin(torch.tensor(phase))).to(self.device)
        
        # Immune centers (can have multiple locations)
        for center, info in self.immune_centers.items():
            positions = info['pos'] if isinstance(info['pos'], list) else [info['pos']]
            
            for px, py in positions:
                envelope = torch.exp(-15 * ((X - px)**2 + (Y - py)**2))
                phase = 2 * np.pi * info['freq'] * 0.1
                
                # Cytokine level modulates amplitude
                amplitude = self.cytokine_levels[center] * 0.7
                # Create complex exponential using cos + i*sin
                self.field += amplitude * envelope * torch.complex(torch.cos(torch.tensor(phase)), torch.sin(torch.tensor(phase))).to(self.device)
    
    def apply_circadian_modulation(self):
        """Apply 24-hour circadian rhythm to hormone production."""
        # Different glands peak at different times
        circadian_peaks = {
            'pineal': np.pi,        # Peaks at night (melatonin)
            'adrenals': np.pi/2,    # Morning cortisol peak
            'pituitary': np.pi/3,   # Early morning
            'thyroid': 2*np.pi/3,   # Mid-morning
            'gonads': 0,            # Variable
        }
        
        for gland in self.glands:
            if gland in circadian_peaks:
                peak_phase = circadian_peaks[gland]
                # Sinusoidal modulation
                modulation = 0.5 + 0.5 * np.cos(self.circadian_phase - peak_phase)
                self.hormone_levels[gland] = 0.3 + 0.7 * modulation
    
    def propagate_coupling(self):
        """Propagate coupling between endocrine and immune systems."""
        # Apply known couplings
        for (source, target), strength in self.coupling_matrix.items():
            if source in self.glands and target == 'all':
                # Pituitary affects everything
                source_level = self.hormone_levels[source]
                for immune in self.immune_centers:
                    self.cytokine_levels[immune] += strength * source_level * 0.01
            
            elif source in self.glands and target in self.immune_centers:
                # Direct gland-to-immune coupling
                source_level = self.hormone_levels[source]
                self.cytokine_levels[target] += strength * source_level * 0.01
        
        # Immune feedback to endocrine (inflammation affects hormones)
        total_inflammation = sum(self.cytokine_levels.values()) / len(self.cytokine_levels)
        
        # Inflammation suppresses sex hormones, boosts stress hormones
        self.hormone_levels['adrenals'] += 0.02 * total_inflammation
        self.hormone_levels['gonads'] *= (1 - 0.1 * total_inflammation)
        
        # Keep levels in range [0, 1]
        for gland in self.hormone_levels:
            self.hormone_levels[gland] = np.clip(self.hormone_levels[gland], 0.1, 1.0)
        for center in self.cytokine_levels:
            self.cytokine_levels[center] = np.clip(self.cytokine_levels[center], 0.1, 1.0)
    
    def add_stress_response(self, stress_intensity=0.5):
        """Model stress response - HPA axis activation."""
        self.stress_level = stress_intensity
        
        # Immediate effects
        self.hormone_levels['adrenals'] += stress_intensity * 0.5  # Cortisol surge
        self.hormone_levels['pituitary'] += stress_intensity * 0.3  # ACTH release
        
        # Suppress immunity
        for center in self.immune_centers:
            self.cytokine_levels[center] *= (1 - stress_intensity * 0.3)
    
    def add_infection(self, x, y, severity=0.5):
        """Add infection site - triggers immune response."""
        self.infection_sites.append({'pos': (x, y), 'severity': severity})
        
        # Nearest lymph nodes activate
        x_tensor = torch.tensor(x, device=self.device)
        y_tensor = torch.tensor(y, device=self.device)
        
        # Boost nearby immune centers
        for center, info in self.immune_centers.items():
            positions = info['pos'] if isinstance(info['pos'], list) else [info['pos']]
            
            for px, py in positions:
                distance = np.sqrt((x - px)**2 + (y - py)**2)
                if distance < 0.5:  # Nearby
                    self.cytokine_levels[center] += severity * (0.5 - distance)
    
    def evolve_one_step(self, dt=0.1):
        """Evolve the coupled endocrine-immune field."""
        # Update circadian phase
        self.circadian_phase += dt * 2 * np.pi / 24  # Full cycle = 24 units
        
        # Apply circadian modulation
        self.apply_circadian_modulation()
        
        # Propagate couplings
        self.propagate_coupling()
        
        # Decay stress
        self.stress_level *= 0.99
        
        # Update field
        self.field *= 0.95  # Decay
        self.initialize_gland_fields()  # Regenerate based on current levels
        
        # Add infection perturbations
        x = torch.linspace(-1, 1, self.size, device=self.device)
        y = torch.linspace(-1, 1, self.size, device=self.device)
        X, Y = torch.meshgrid(x, y, indexing='ij')
        
        for infection in self.infection_sites:
            ix, iy = infection['pos']
            severity = infection['severity']
            
            # Infection creates chaotic field perturbation
            infection_field = severity * torch.exp(-20 * ((X - ix)**2 + (Y - iy)**2))
            noise = torch.randn_like(self.field) * 0.1
            self.field += infection_field.unsqueeze(0).unsqueeze(0) * (1 + noise)
            
            # Decay infection
            infection['severity'] *= 0.98
        
        # Remove healed infections
        self.infection_sites = [i for i in self.infection_sites if i['severity'] > 0.01]
    
    def calculate_system_coherence(self):
        """Calculate overall endocrine-immune coherence."""
        # Fourier transform to get frequency spectrum
        spectrum = torch.fft.fft2(self.field[0, 0])
        power_spectrum = torch.abs(spectrum)**2
        
        # Coherence is ratio of peak power to total power
        total_power = power_spectrum.sum()
        peak_power = power_spectrum.max()
        
        coherence = (peak_power / total_power).item()
        return coherence

# Create the unified field
endo_immune = EndocrineImmuneField(size=256, device=device)
endo_immune.initialize_gland_fields()

print("✓ Endocrine-Immune field initialized")
print(f"  {len(endo_immune.glands)} endocrine glands")
print(f"  {len(endo_immune.immune_centers)} immune centers")
print(f"  {len(endo_immune.coupling_matrix)} coupling relationships")
```

```python
def simulate_disease_state(disease_type='Healthy', duration=100):
    """Simulate different disease states affecting endocrine-immune balance."""
    
    # Create fresh system
    disease_field = EndocrineImmuneField(size=128, device=device)
    disease_field.initialize_gland_fields()
    
    # Disease-specific modifications
    if disease_type == 'Autoimmune':
        # Overactive immune system
        for center in disease_field.immune_centers:
            disease_field.cytokine_levels[center] *= 2.0
        # Stressed adrenals trying to compensate
        disease_field.hormone_levels['adrenals'] *= 1.5
        
    elif disease_type == 'Chronic Stress':
        # Persistent HPA activation
        disease_field.stress_level = 0.8
        disease_field.hormone_levels['adrenals'] = 0.9
        # Suppressed immunity
        for center in disease_field.immune_centers:
            disease_field.cytokine_levels[center] *= 0.5
            
    elif disease_type == 'Thyroid Disorder':
        # Hypothyroid - low metabolism
        disease_field.hormone_levels['thyroid'] = 0.2
        # Affects everything
        disease_field.hormone_levels['pituitary'] *= 1.3  # Trying to stimulate
        
    elif disease_type == 'Diabetes':
        # Pancreatic dysfunction
        disease_field.hormone_levels['pancreas'] = 0.3
        # Chronic inflammation
        for center in disease_field.immune_centers:
            disease_field.cytokine_levels[center] *= 1.2
    
    # Run simulation
    coherence_trace = []
    for t in range(duration):
        disease_field.evolve_one_step(dt=0.1)
        coherence_trace.append(disease_field.calculate_system_coherence())
    
    # Visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # Field state
    field_mag = torch.abs(disease_field.field[0, 0]).cpu().numpy()
    im = ax1.imshow(field_mag, cmap='twilight', extent=[-1, 1, -1, 1])
    ax1.set_title(f'{disease_type} - Field State', fontsize=14)
    ax1.set_xlabel('X')
    ax1.set_ylabel('Y')
    plt.colorbar(im, ax=ax1)
    
    # System balance
    categories = ['Endocrine\nBalance', 'Immune\nBalance', 'System\nCoherence']
    
    # Calculate balance metrics
    endocrine_balance = np.std([disease_field.hormone_levels[g] for g in disease_field.glands])
    immune_balance = np.std([disease_field.cytokine_levels[c] for c in disease_field.immune_centers])
    final_coherence = coherence_trace[-1]
    
    # Invert standard deviations (lower is better)
    endocrine_score = 1 / (1 + endocrine_balance)
    immune_score = 1 / (1 + immune_balance)
    coherence_score = final_coherence * 10  # Scale up
    
    values = [endocrine_score, immune_score, coherence_score]
    colors = ['red', 'blue', 'purple']
    
    bars = ax2.bar(categories, values, color=colors, alpha=0.7)
    ax2.set_ylabel('Balance Score', fontsize=12)
    ax2.set_title(f'{disease_type} - System Balance', fontsize=14)
    ax2.set_ylim(0, 1)
    
    # Add value labels
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                f'{value:.2f}', ha='center', va='bottom')
    
    # Add healthy reference line
    ax2.axhline(y=0.7, color='green', linestyle='--', alpha=0.5, label='Healthy Range')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
    return disease_field, coherence_trace

# Interactive widget
disease_widget = widgets.Dropdown(
    options=['Healthy', 'Autoimmune', 'Chronic Stress', 'Thyroid Disorder', 'Diabetes'],
    value='Healthy',
    description='Disease:'
)

duration_widget = widgets.IntSlider(
    value=100,
    min=50,
    max=200,
    step=10,
    description='Duration:'
)

interact(simulate_disease_state, disease_type=disease_widget, duration=duration_widget)
```
