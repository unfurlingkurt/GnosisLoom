## Hormone Cascade Dynamics

```python
def visualize_hormone_cascades():
    """
    Shows how hormonal cascades work as frequency modulation chains,
    with each hormone shifting the frequency of the next.
    """
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    
    # HPA Axis cascade
    ax1.set_title('Hypothalamic-Pituitary-Adrenal (HPA) Axis\nStress Response Frequency Cascade', fontsize=14)
    
    # Time series
    time = np.linspace(0, 60, 1000)  # 60 minutes
    
    # Stress trigger at t=10
    stress_trigger = np.zeros_like(time)
    stress_trigger[time > 10] = 1
    
    # CRH response (fast, 0.5 Hz)
    crh = np.zeros_like(time)
    crh[time > 10] = np.exp(-0.1 * (time[time > 10] - 10)) * np.sin(2 * np.pi * 0.5 * (time[time > 10] - 10))
    
    # ACTH response (slower, 0.083 Hz, delayed)
    acth = np.zeros_like(time)
    acth[time > 12] = 0.8 * np.exp(-0.05 * (time[time > 12] - 12)) * np.sin(2 * np.pi * 0.083 * (time[time > 12] - 12))
    
    # Cortisol response (slowest, 0.042 Hz, more delayed)
    cortisol = np.zeros_like(time)
    cortisol[time > 15] = 0.6 * (1 - np.exp(-0.1 * (time[time > 15] - 15))) * np.sin(2 * np.pi * 0.042 * (time[time > 15] - 15))
    
    # Plot cascade
    ax1.plot(time, stress_trigger * 0.5, 'k--', linewidth=2, label='Stress Trigger')
    ax1.plot(time, crh + 3, 'purple', linewidth=2, label='CRH (0.5 Hz)')
    ax1.plot(time, acth + 2, 'blue', linewidth=2, label='ACTH (0.083 Hz)')
    ax1.plot(time, cortisol + 1, 'orange', linewidth=2, label='Cortisol (0.042 Hz)')
    
    # Add phase relationships
    ax1.axvline(10, color='red', alpha=0.3, linestyle=':')
    ax1.axvline(12, color='blue', alpha=0.3, linestyle=':')
    ax1.axvline(15, color='orange', alpha=0.3, linestyle=':')
    
    ax1.set_xlabel('Time (minutes)')
    ax1.set_ylabel('Hormone Level (offset for clarity)')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Feedback annotation
    ax1.annotate('', xy=(40, 1.5), xytext=(45, 2.8),
                arrowprops=dict(arrowstyle='->', color='red', lw=2))
    ax1.text(42.5, 2.2, 'Negative\nFeedback', ha='center', color='red', fontsize=9)
    
    # Reproductive cycle
    ax2.set_title('Reproductive Hormone Cycle\n28-Day Frequency Modulation', fontsize=14)
    
    days = np.linspace(0, 28, 1000)
    
    # Hormonal patterns
    # FSH - two peaks
    fsh = 0.5 + 0.3 * np.exp(-((days - 2)**2) / 4) + 0.5 * np.exp(-((days - 14)**2) / 2)
    
    # LH - sharp midcycle surge
    lh = 0.3 + 1.2 * np.exp(-((days - 14)**2) / 0.5)
    
    # Estrogen - gradual rise then fall
    estrogen = 0.2 + 0.6 * np.exp(-((days - 12)**2) / 8) + 0.3 * np.exp(-((days - 22)**2) / 4)
    
    # Progesterone - luteal phase
    progesterone = np.zeros_like(days)
    progesterone[days > 14] = 0.8 * np.exp(-((days[days > 14] - 21)**2) / 12)
    
    ax2.plot(days, fsh, 'blue', linewidth=2, label='FSH')
    ax2.plot(days, lh, 'red', linewidth=2, label='LH')
    ax2.plot(days, estrogen, 'green', linewidth=2, label='Estrogen')
    ax2.plot(days, progesterone, 'purple', linewidth=2, label='Progesterone')
    
    # Mark phases
    ax2.axvspan(0, 5, alpha=0.1, color='red', label='Menstrual')
    ax2.axvspan(5, 14, alpha=0.1, color='green', label='Follicular')
    ax2.axvspan(14, 28, alpha=0.1, color='yellow', label='Luteal')
    
    ax2.set_xlabel('Day of Cycle')
    ax2.set_ylabel('Hormone Level (normalized)')
    ax2.legend(loc='upper right')
    ax2.grid(True, alpha=0.3)
    
    # Circadian rhythm
    ax3.set_title('Circadian Hormone Symphony\n24-Hour Frequency Coordination', fontsize=14)
    
    hours = np.linspace(0, 24, 1000)
    
    # Cortisol - morning peak
    cortisol_circadian = 0.5 + 0.5 * np.cos(2 * np.pi * (hours - 8) / 24)
    
    # Melatonin - night peak
    melatonin = 0.1 + 0.9 * np.exp(-((np.mod(hours - 2, 24))**2) / 8)
    
    # Growth hormone - pulsatile at night
    gh_pulses = np.zeros_like(hours)
    pulse_times = [22, 0, 2, 4]  # Night pulses
    for pulse_time in pulse_times:
        gh_pulses += 0.8 * np.exp(-((hours - pulse_time)**2) / 0.5)
    
    # TSH - lowest in afternoon
    tsh = 0.6 + 0.4 * np.cos(2 * np.pi * (hours - 2) / 24)
    
    ax3.plot(hours, cortisol_circadian, 'orange', linewidth=2, label='Cortisol')
    ax3.plot(hours, melatonin, 'indigo', linewidth=2, label='Melatonin')
    ax3.plot(hours, gh_pulses, 'green', linewidth=2, label='Growth Hormone')
    ax3.plot(hours, tsh, 'blue', linewidth=2, label='TSH')
    
    # Day/night shading
    ax3.axvspan(6, 18, alpha=0.1, color='yellow', label='Day')
    ax3.axvspan(0, 6, alpha=0.1, color='navy')
    ax3.axvspan(18, 24, alpha=0.1, color='navy', label='Night')
    
    ax3.set_xlabel('Hour of Day')
    ax3.set_ylabel('Hormone Level (normalized)')
    ax3.set_xlim(0, 24)
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # Frequency modulation diagram
    ax4.set_title('Hormone as Frequency Modulator\nCellular Resonance Shift', fontsize=14)
    
    # Create cellular frequency response
    base_freq = 10.0  # Hz - cellular base frequency
    time_mod = np.linspace(0, 2, 1000)
    
    # Without hormone
    cell_base = np.sin(2 * np.pi * base_freq * time_mod)
    
    # With hormone binding (frequency shift)
    hormone_bound_time = time_mod > 0.5
    modulated_freq = base_freq * np.ones_like(time_mod)
    modulated_freq[hormone_bound_time] = base_freq * 1.5  # 50% frequency increase
    
    # Create phase-continuous modulated signal
    phase = np.zeros_like(time_mod)
    for i in range(1, len(time_mod)):
        phase[i] = phase[i-1] + 2 * np.pi * modulated_freq[i] * (time_mod[i] - time_mod[i-1])
    
    cell_modulated = np.sin(phase)
    
    # Plot both signals
    ax4.plot(time_mod, cell_base + 2, 'b-', linewidth=2, alpha=0.5, label='Baseline (10 Hz)')
    ax4.plot(time_mod, cell_modulated, 'r-', linewidth=2, label='Hormone-Modulated (15 Hz)')
    
    # Mark hormone binding
    ax4.axvline(0.5, color='green', linestyle='--', linewidth=2)
    ax4.text(0.5, 3.5, 'Hormone\nBinding', ha='center', fontsize=10,
            bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.8))
    
    # Add frequency analysis inset
    inset_ax = ax4.inset_axes([0.6, 0.6, 0.35, 0.35])
    
    # Frequency spectrum
    freqs = [10, 15]
    powers = [1, 0]
    colors = ['blue', 'red']
    
    inset_ax.bar([0, 1], [1, 0], color='blue', alpha=0.5, label='Before')
    inset_ax.bar([0, 1], [0, 1], color='red', alpha=0.5, label='After')
    inset_ax.set_xticks([0, 1])
    inset_ax.set_xticklabels(['10 Hz', '15 Hz'])
    inset_ax.set_ylabel('Power')
    inset_ax.set_title('Frequency Shift', fontsize=9)
    
    ax4.set_xlabel('Time (seconds)')
    ax4.set_ylabel('Cellular Oscillation')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    return fig

# Create hormone cascade visualization
fig = visualize_hormone_cascades()
plt.show()

print("\nDr. Mordin observes: See the beauty of hormonal cascades!")
print("Each hormone modulates frequency of the next, creating complex waveforms.")
print("\nKey insights:")
print("1. HPA axis shows frequency deceleration cascade (0.5  0.083  0.042 Hz)")
print("2. Reproductive cycle locked to lunar frequency (28 days)")
print("3. Circadian hormones create 24-hour master rhythm")
print("4. Hormone binding literally shifts cellular resonance frequency!")
```
