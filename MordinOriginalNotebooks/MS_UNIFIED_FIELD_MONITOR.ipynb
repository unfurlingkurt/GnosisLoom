## Section 3: Interactive System Explorer

```python
# Create interactive system explorer
def create_system_explorer():
    # Dropdown for system selection
    system_dropdown = widgets.Dropdown(
        options=list(SYSTEM_STATUS.keys()),
        value='Neural',
        description='System:',
        style={'description_width': 'initial'}
    )
    
    # Output area
    output = widgets.Output()
    
    def update_display(change):
        with output:
            clear_output(wait=True)
            
            system = change['new']
            data = SYSTEM_STATUS[system]
            
            fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))
            
            # Completion gauge
            ax1.clear()
            completion = data['completion']
            color = 'green' if completion >= 85 else 'yellow' if completion >= 50 else 'red'
            
            # Create circular progress
            theta = np.linspace(0, 2*np.pi*completion/100, 100)
            r = 1
            x = r * np.cos(theta)
            y = r * np.sin(theta)
            
            ax1.plot(x, y, color=color, linewidth=20)
            ax1.text(0, 0, f"{completion}%", ha='center', va='center', 
                    fontsize=40, fontweight='bold')
            ax1.set_xlim(-1.5, 1.5)
            ax1.set_ylim(-1.5, 1.5)
            ax1.set_aspect('equal')
            ax1.set_title(f'{system} System Completion', fontsize=16)
            ax1.axis('off')
            
            # Frequency spectrum
            ax2.clear()
            if data['frequencies']:
                freqs = data['frequencies']
                ax2.scatter(range(len(freqs)), freqs, s=100, c='blue', edgecolors='white', linewidth=2)
                ax2.set_yscale('log')
                ax2.set_xlabel('Frequency Index')
                ax2.set_ylabel('Frequency (Hz)')
                ax2.set_title('System Frequencies')
                ax2.grid(True, alpha=0.3)
                
                # Add frequency labels
                for i, freq in enumerate(freqs):
                    if freq < 1:
                        label = f"{freq:.3f} Hz"
                    elif freq < 1000:
                        label = f"{freq:.1f} Hz"
                    else:
                        label = f"{freq:.0e} Hz"
                    ax2.text(i, freq*1.5, label, ha='center', fontsize=8)
            
            # Notebook coverage
            ax3.clear()
            ax3.text(0.1, 0.9, "Notebooks:", fontsize=14, fontweight='bold', 
                    transform=ax3.transAxes)
            
            notebook_text = "\n".join(f"• {nb}" for nb in data['notebooks'])
            ax3.text(0.1, 0.7, notebook_text, fontsize=12, transform=ax3.transAxes,
                    verticalalignment='top')
            
            ax3.text(0.1, 0.4, "Stellar Resonance:", fontsize=14, fontweight='bold',
                    transform=ax3.transAxes)
            
            stellar_text = "\n".join(f"⭐ {star}" for star in data['stellar_resonance'])
            ax3.text(0.1, 0.3, stellar_text, fontsize=12, transform=ax3.transAxes,
                    verticalalignment='top', color='gold')
            
            ax3.axis('off')
            
            # Missing components
            ax4.clear()
            ax4.text(0.1, 0.9, "Missing Components:", fontsize=14, fontweight='bold',
                    transform=ax4.transAxes, color='red')
            
            missing_text = "\n".join(f"• {item}" for item in data['missing'])
            ax4.text(0.1, 0.8, missing_text, fontsize=11, transform=ax4.transAxes,
                    verticalalignment='top', color='orange')
            
            ax4.axis('off')
            
            plt.tight_layout()
            plt.show()
    
    # Connect the update function
    system_dropdown.observe(update_display, names='value')
    
    # Initial display
    update_display({'new': system_dropdown.value})
    
    # Layout
    return widgets.VBox([system_dropdown, output])

# Create and display the explorer
explorer = create_system_explorer()
display(explorer)
```

```python
# Calculate consciousness coherence score
def calculate_consciousness_coherence():
    """
    Calculate overall consciousness coherence based on:
    - System completion
    - Frequency harmony
    - Cross-system coupling
    - Stellar resonance alignment
    """
    
    # Component scores
    scores = {}
    
    # 1. Completion score (weighted by system importance)
    weights = {
        'Neural': 1.5,
        'Metabolic': 1.3,
        'Endocrine': 1.2,
        'Cardiovascular': 1.2,
        'Immune': 1.1,
        'Sensory': 1.0,
        'Connective': 1.0,
        'Digestive': 0.9,
        'Respiratory': 0.9,
        'Musculoskeletal': 0.8,
        'Integumentary': 0.7,
        'Reproductive': 0.7,
        'Urinary': 0.6
    }
    
    weighted_completion = sum(SYSTEM_STATUS[sys]['completion'] * weights[sys] 
                             for sys in SYSTEM_STATUS) / sum(weights.values())
    scores['Completion'] = weighted_completion
    
    # 2. Frequency harmony score (check for harmonic relationships)
    all_freqs = []
    for data in SYSTEM_STATUS.values():
        all_freqs.extend(data['frequencies'])
    
    harmonic_pairs = 0
    total_pairs = 0
    for i, f1 in enumerate(all_freqs):
        for f2 in all_freqs[i+1:]:
            if min(f1, f2) > 0:
                ratio = max(f1, f2) / min(f1, f2)
                # Check for harmonic ratios
                for h in [2, 3, 4, 5, 8, 1.618]:  # Including golden ratio
                    if abs(ratio - h) < 0.1 or abs(ratio - 1/h) < 0.1:
                        harmonic_pairs += 1
                        break
                total_pairs += 1
    
    harmony_score = (harmonic_pairs / total_pairs * 100) if total_pairs > 0 else 0
    scores['Harmony'] = harmony_score
    
    # 3. Cross-system coupling score
    coupled_systems = 0
    for sys1 in SYSTEM_STATUS:
        for sys2 in SYSTEM_STATUS:
            if sys1 != sys2:
                # Check if systems share stellar resonance
                stars1 = set(SYSTEM_STATUS[sys1]['stellar_resonance'])
                stars2 = set(SYSTEM_STATUS[sys2]['stellar_resonance'])
                if stars1 & stars2:  # Intersection
                    coupled_systems += 1
    
    max_coupling = len(SYSTEM_STATUS) * (len(SYSTEM_STATUS) - 1)
    coupling_score = (coupled_systems / max_coupling * 100) if max_coupling > 0 else 0
    scores['Coupling'] = coupling_score
    
    # 4. Myelic fluid factor (bonus for consciousness discovery)
    myelic_bonus = 20  # 20% bonus for discovering consciousness substrate
    scores['Myelic'] = myelic_bonus
    
    # Calculate final coherence
    coherence = np.mean(list(scores.values()))
    
    return coherence, scores

# Visualize consciousness coherence
coherence, component_scores = calculate_consciousness_coherence()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Coherence meter
ax1.clear()
theta = np.linspace(0, np.pi, 100)  # Half circle
r_outer = 1
r_inner = 0.7

# Background arc
x_outer = r_outer * np.cos(theta)
y_outer = r_outer * np.sin(theta)
ax1.fill_between(x_outer, 0, y_outer, color='lightgray', alpha=0.3)

# Coherence arc
theta_coherence = np.linspace(0, np.pi * coherence/100, 100)
x_coherence = r_outer * np.cos(theta_coherence)
y_coherence = r_outer * np.sin(theta_coherence)

# Color based on score
if coherence >= 80:
    color = 'green'
    status = 'CRYSTALLIZED'
elif coherence >= 60:
    color = 'yellow'
    status = 'EMERGING'
else:
    color = 'red'
    status = 'DEVELOPING'

ax1.fill_between(x_coherence, 0, y_coherence, color=color, alpha=0.7)

# Add text
ax1.text(0, -0.2, f"{coherence:.1f}%", ha='center', va='center', 
         fontsize=48, fontweight='bold')
ax1.text(0, -0.4, status, ha='center', va='center', 
         fontsize=24, color=color, fontweight='bold')

# Add scale markers
for angle, label in [(0, '0'), (np.pi/4, '25'), (np.pi/2, '50'), 
                     (3*np.pi/4, '75'), (np.pi, '100')]:
    x = 1.1 * np.cos(angle)
    y = 1.1 * np.sin(angle)
    ax1.text(x, y, label, ha='center', va='center', fontsize=10)

ax1.set_xlim(-1.5, 1.5)
ax1.set_ylim(-0.5, 1.5)
ax1.set_aspect('equal')
ax1.set_title('Consciousness Coherence Score', fontsize=20, pad=20)
ax1.axis('off')

# Component breakdown
ax2.clear()
components = list(component_scores.keys())
values = list(component_scores.values())
colors_comp = ['blue', 'green', 'orange', 'purple']

bars = ax2.bar(components, values, color=colors_comp, edgecolor='white', linewidth=2)

# Add value labels
for bar, val in zip(bars, values):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, height + 1,
             f'{val:.1f}%', ha='center', va='bottom', fontsize=12, fontweight='bold')

ax2.set_ylim(0, 100)
ax2.set_ylabel('Score (%)', fontsize=14)
ax2.set_title('Coherence Component Scores', fontsize=16)
ax2.grid(True, axis='y', alpha=0.3)

# Add description
desc_text = (
    "Consciousness Coherence integrates:\n"
    "• System completion (weighted)\n"
    "• Frequency harmony\n"
    "• Cross-system coupling\n"
    "• Myelic fluid discovery bonus"
)
ax2.text(0.02, 0.98, desc_text, transform=ax2.transAxes, 
         fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='black', alpha=0.8))

plt.tight_layout()
plt.show()

print("\nConsciousness Coherence Analysis:")
print(f"- Overall Coherence: {coherence:.1f}% ({status})")
print(f"- Weighted Completion: {component_scores['Completion']:.1f}%")
print(f"- Frequency Harmony: {component_scores['Harmony']:.1f}%")
print(f"- System Coupling: {component_scores['Coupling']:.1f}%")
print(f"- Myelic Bonus: {component_scores['Myelic']:.1f}%")
print("\n'Consciousness crystallizes when coherence exceeds 80%!'")
```

```python
# Create research priority matrix
def calculate_research_priority():
    priorities = []
    
    for system, data in SYSTEM_STATUS.items():
        # Factors for priority calculation
        incompleteness = 100 - data['completion']
        
        # System importance (same weights as before)
        importance = {
            'Neural': 1.5, 'Metabolic': 1.3, 'Endocrine': 1.2,
            'Cardiovascular': 1.2, 'Immune': 1.1, 'Sensory': 1.0,
            'Connective': 1.0, 'Digestive': 0.9, 'Respiratory': 0.9,
            'Musculoskeletal': 0.8, 'Integumentary': 0.7,
            'Reproductive': 0.7, 'Urinary': 0.6
        }.get(system, 1.0)
        
        # Number of missing components
        n_missing = len(data['missing'])
        
        # Calculate priority score
        priority_score = incompleteness * importance * (1 + n_missing/10)
        
        priorities.append({
            'system': system,
            'completion': data['completion'],
            'priority_score': priority_score,
            'missing_count': n_missing,
            'importance': importance
        })
    
    # Sort by priority score
    priorities.sort(key=lambda x: x['priority_score'], reverse=True)
    
    return priorities

# Visualize priority matrix
priorities = calculate_research_priority()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Priority ranking
ax1.clear()
top_n = 8  # Show top 8 priorities
top_priorities = priorities[:top_n]

systems = [p['system'] for p in top_priorities]
scores = [p['priority_score'] for p in top_priorities]
completions = [p['completion'] for p in top_priorities]

# Color based on completion
colors = ['red' if c < 50 else 'orange' if c < 85 else 'green' for c in completions]

y_pos = np.arange(len(systems))
bars = ax1.barh(y_pos, scores, color=colors, edgecolor='white', linewidth=2)

# Add labels
ax1.set_yticks(y_pos)
ax1.set_yticklabels(systems)
ax1.set_xlabel('Priority Score', fontsize=12)
ax1.set_title('Research Priority Ranking', fontsize=16)

# Add completion percentage on bars
for i, (bar, comp) in enumerate(zip(bars, completions)):
    width = bar.get_width()
    ax1.text(width + 1, bar.get_y() + bar.get_height()/2,
             f'{comp}%', ha='left', va='center', fontsize=10)

# Priority matrix scatter
ax2.clear()
importance_values = [p['importance'] for p in priorities]
incompleteness = [100 - p['completion'] for p in priorities]
missing_counts = [p['missing_count'] for p in priorities]

# Bubble chart
scatter = ax2.scatter(importance_values, incompleteness, 
                     s=[m*100 for m in missing_counts],
                     c=incompleteness, cmap='RdYlGn_r',
                     edgecolors='white', linewidth=2, alpha=0.7)

# Add labels for high priority systems
for p in priorities:
    if p['priority_score'] > 30:
        ax2.annotate(p['system'][:4], 
                    (p['importance'], 100 - p['completion']),
                    xytext=(5, 5), textcoords='offset points',
                    fontsize=9, fontweight='bold')

ax2.set_xlabel('System Importance', fontsize=12)
ax2.set_ylabel('Incompleteness (%)', fontsize=12)
ax2.set_title('Priority Matrix: Importance vs Incompleteness', fontsize=16)
ax2.grid(True, alpha=0.3)

# Add quadrants
ax2.axhline(50, color='gray', linestyle='--', alpha=0.5)
ax2.axvline(1.0, color='gray', linestyle='--', alpha=0.5)

# Quadrant labels
ax2.text(0.7, 75, 'Low Priority\nIncomplete', ha='center', va='center',
         fontsize=10, alpha=0.5, style='italic')
ax2.text(1.3, 75, 'HIGH PRIORITY\nIncomplete', ha='center', va='center',
         fontsize=11, alpha=0.7, weight='bold', color='red')
ax2.text(0.7, 25, 'Low Priority\nComplete', ha='center', va='center',
         fontsize=10, alpha=0.5, style='italic')
ax2.text(1.3, 25, 'Maintenance\nComplete', ha='center', va='center',
         fontsize=10, alpha=0.5, style='italic')

# Colorbar
cbar = plt.colorbar(scatter, ax=ax2)
cbar.set_label('Incompleteness (%)', fontsize=10)

plt.tight_layout()
plt.show()

# Print research roadmap
print("\nResearch Priority Roadmap:")
print("\nHIGH PRIORITY (Next notebooks):")
for i, p in enumerate(priorities[:4]):
    print(f"{i+1}. {p['system']} ({p['completion']}% complete)")
    print(f"   Priority Score: {p['priority_score']:.1f}")
    print(f"   Missing: {p['missing_count']} components")

print("\nMEDIUM PRIORITY:")
for i, p in enumerate(priorities[4:7]):
    print(f"{i+5}. {p['system']} ({p['completion']}% complete)")

print("\n'Focus on Respiratory and Urinary - the breath and flow of consciousness!'")
```
